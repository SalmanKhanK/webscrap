```python
import requests as re
from bs4 import BeautifulSoup
```

```python
for i in range(1,32): 
    url="https://www.pakwheels.com/used-cars/search/-/ct_karachi/ca_afohs/ca_air-port/ca_cantt/ca_jinnah-avenue--2/ca_malir/ca_model-colony/ca_national-highway/ca_others-0/ca_rafah-e-aam/ca_safoora/ca_shah-faisal-colony/ca_shah-latif-town/ca_shahra-e-faisal/ca_university-road-1/?page="+str(i)+""
    r=re.get(url)
    a=0;
    #step1 get the html
    htmlContent=r.content
    soup=BeautifulSoup(htmlContent,'html.parser')
    phone=soup.find_all('span',class_='generic-green fs16')
    for cell in phone:
        print(cell.get_text())
        a=a+1
print("Total",a)   
```
